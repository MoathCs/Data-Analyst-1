```python

# 1 Columns with Non-Essential Textual Information , Replace null values with a placeholder like "Unknown"
df['homepage'].fillna("Unknown", inplace=True)
df['tagline'].fillna("Unknown", inplace=True)

# 2 : Columns with Descriptive Text
df['overview'].fillna("No overview available", inplace=True)

# 3 drop cuz could impact to  analysis result
df.dropna(subset=['release_date'], inplace=True)

#4 Numeric Data fill as mean 

df['runtime'].fillna(df['runtime'].mean(), inplace=True)

```

```python
import matplotlib.pyplot as plt
# Budget distribution
df['budget'].hist(bins=30)
plt.title("Distribution of Budget")
plt.xlabel("Budget")
plt.ylabel("Frequency")
plt.show()

```

This histogram represents the distribution of movie budgets.

X-Axis (Budget): Shows the budget of each movie, measured in dollars. The budget values range from 0 up to around 3.5 * 10^8 (350 million).

Y-Axis (Frequency): Indicates how many movies fall within each budget range.

From this plot, we can see that most movies have a relatively low budget, with a sharp decrease in the frequency as the budget increases. This suggests that high-budget movies are less common. Further analysis can explore how budget impacts revenue and other factors.

```python
# Revenue distribution
df['revenue'].hist(bins=30)
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

This histogram represents the distribution of movie revenues.

X-Axis (Revenue): Displays the revenue generated by each movie, measured in dollars. Revenue values range from 0 up to around 2.5 * 10^9 (2.5 billion).

Y-Axis (Frequency): Indicates how many movies fall within each revenue range.

From this plot, it's evident that most movies generate relatively low revenue, with the frequency dropping sharply as revenue increases. High-revenue movies are relatively rare. This suggests that while a few movies earn very high revenue, the majority earn significantly less. Further analysis will help examine how different factors, like budget, influence revenue distribution.








```python
# Plots the relationship between two variables as a scatter plot.

def scatter_chart(arg1,arg2):  

    plt.figure(figsize=(12,10))  

    plt.scatter(x=df[arg1],y=df[arg2],c='blue') #when you plot this chart instead ot [arg1] > "column_name"  

    plt.title(f"The Relation Betweem {arg1.title()} and {arg2.title()}",fontsize=14) # instead of [arg2] > "column_name"  

    plt.xlabel(f"{arg1.title()}",fontsize=12)  

    plt.ylabel(f"{arg2.title()}",fontsize=14)  

    plt.show()  
```

```python
# #Relationship between Budget and revenue.

scatter_chart('budget' , 'revenue')
```

There is some positive correlation between budget and revenue, but the relationship is not very strong. This will be examined further in later analysis.

X-Axis (Budget): Represents the independent variable, budget. This axis shows the budget allocated to each instance (e.g., movie) in millions, ranging from 0 up to around 350 million.

Y-Axis (Revenue): Represents the dependent variable, revenue. This axis shows the revenue generated by each instance, with values reaching up to about 2.5 billion.

## Conclusions
Summary of Findings
Based on the exploratory data analysis (EDA) conducted on the dataset, several factors were identified as influencing movie revenue. Below is a summary of the key findings:

Budget: There is a strong positive correlation between a movie's budget and its revenue. Larger budgets tend to be associated with higher revenues, likely due to better production quality, larger marketing campaigns, and more extensive distribution.

Popularity: Popularity also shows a moderate positive correlation with revenue. This suggests that movies that are more popular—perhaps through better marketing, high anticipation, or social media presence—generally earn higher revenues.

Vote Count and Vote Average: There is a moderate positive relationship between vote count and vote average with revenue. Movies with more votes and higher ratings tend to earn more, indicating that both audience and critical reception are important, although not as strongly as budget and popularity.

Genre: The genre of the movie also seems to affect its revenue, though this relationship requires further analysis. Some genres like action or adventure likely perform better at the box office compared to others like drama or romance.

Limitations of the Analysis
Missing Data and Assumptions: The dataset had missing values for certain columns (e.g., homepage, tagline), which were filled with mean values or dropped. However, the impact of these missing values on the analysis was not fully addressed. The absence of certain key data (such as detailed marketing spend or audience demographics) might also limit the depth of the analysis.

Correlation Does Not Imply Causation: While there are strong correlations between factors like budget and revenue, it is important to note that correlation does not imply causation. The observed relationships could be influenced by other factors not captured in the dataset, such as external market conditions, competition, or movie reviews. Without performing statistical tests (e.g., regression analysis), we cannot draw conclusions about causality.

Limited Genre Exploration: While genre was examined briefly, a more detailed analysis could be conducted to determine how different genres influence revenue across various regions or over time. This would help identify more nuanced trends.

Conclusion
From this analysis, we conclude that the budget and popularity of a movie are the most influential factors on its revenue. Other factors, such as vote count and vote average, also play a role, but their influence is not as strong. This analysis highlights the importance of financial investment and public perception in determining a movie's success at the box office. However, further research, particularly into marketing efforts and genre-specific trends, would provide a more complete picture of the factors driving revenue.
